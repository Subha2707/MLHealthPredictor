# **Column-wise Description of the Dataset**

The dataset contains detailed hospital and patient information. Each record represents one patient and includes demographic, medical, hospital, financial, and outcome-related attributes.

# Patient_ID
 is a unique number assigned to each patient for identification purposes. Patient_Name represents the name of the patient and is mainly used for identification rather than analysis.
# Patient_Name

Name of the patients
# Age
indicates the age of the patient in years and is an important factor in determining health risk and treatment cost. Gender specifies whether the patient is male or female. City shows the location of the patient.

# Height_cm and Weight_kg
 represent the physical measurements of the patient and can be used to calculate Body Mass Index (BMI), which helps assess overall body condition.

# City
The City from where the patient belongs

# Disease
specifies the medical condition diagnosed in the patient, such as Diabetes, Heart Disease, Asthma, or Flu.

# Pulse_Rate
refers to the number of heartbeats per minute and is used to evaluate heart health.

#Oxygen_Level
indicates the oxygen saturation in the blood; lower values may suggest respiratory problems.

# BP (Blood Pressure)
  is used to monitor hypertension or hypotension.
  
# Sugar
represents blood sugar level, which is important for identifying diabetes.

# Chol
 denotes cholesterol level and is associated with heart disease risk.

# Smoke
represents whether the patient is a smoker.

# Alcohol
indicates alcohol consumption habits. These lifestyle factors are important for disease analysis.

# Chronic_Disease
indicates whether the patient has a long-term illness.

# Stress_Level
shows the stress condition of the patient (low, medium, or high), which can influence health.

# ICU_Required
shows whether the patient required intensive care.

# Doctor_Specialization
shows the medical specialty of the doctor treating the patient.

# Appointment_Type
indicates whether the visit was an emergency or follow-up.

# Admit_Type
describes the type of hospital admission (emergency or normal).

# Days
represents the number of days the patient stayed in the hospital.

# Visits
indicates how many times the patient visited the hospital.

# Cost
 represents the total treatment expense incurred by the patient and is the main target variable used in regression analysis.

# Insurance
indicates whether the patient had medical insurance coverage.

# Outcome
describes whether the patient recovered or not after treatment.

# Readmitted
shows whether the patient had to be admitted again after discharge.

# Status
indicates the final status of the patient, such as alive or deceased.



### Data Type Summary

The hospital dataset contains a mix of numerical and categorical variables representing patient details, medical measurements, hospital information, and outcomes.

## Numerical Variables

These variables contain measurable values and are used in statistical and regression analysis:

Age – Patient age in years

 Height_cm – Height in centimeters

Weight_kg – Weight in kilograms

Pulse_Rate – Heart rate per minute

Oxygen_Level – Oxygen saturation percentage

BP – Blood pressure value

Sugar – Blood sugar level

Chol – Cholesterol level

Days – Number of hospital stay days

Visits – Number of hospital visits

Cost – Total treatment cost

These variables are continuous in nature and help in prediction modeling.

## Categorical Variables

These variables represent categories or labels rather than numeric measurements:

Patient_ID – Identifier

Patient_Name – Patient name

Gender – Male/Female

City – Location

Disease – Type of illness

Chronic_Disease – Yes/No

Stress_Level – Low/Medium/High

Smoke – Yes/No

Alcohol – Yes/No

Doctor_Specialization – Medical field

Appointment_Type – Emergency/Follow-up

Admit_Type – Emergency/Normal

ICU_Required – Yes/No

Insurance – Yes/No

Outcome – Recovered/Not Recovered

Readmitted – Yes/No

Status – Alive/Deceased

These variables are converted into numerical form during preprocessing for machine learning models.

```python
num_df.skew()
```

| **Skew Value** | **Meaning**                      | **Distribution Shape**                                                 |
| -------------- | -------------------------------- | ---------------------------------------------------------------------- |
| ≈ 0            | Normal Distribution              | Data is symmetrically distributed around the mean.                     |
| > 0            | Positively Skewed (Right Skewed) | The right tail is longer. Few large values pull the mean to the right. |
| < 0            | Negatively Skewed (Left Skewed)  | The left tail is longer. Few small values pull the mean to the left.   |

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("hospital_dataset_final.csv")

# For better visuals
sns.set_style("whitegrid")
```

```python
plt.figure()
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.figure()
df['Disease'].value_counts().plot(kind='bar')
plt.title("Disease Frequency")
plt.xlabel("Disease Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='ICU_Required', y='Cost', data=df)
plt.title("Cost Comparison: ICU vs Non-ICU Patients")
plt.xlabel("ICU Required")
plt.ylabel("Treatment Cost")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Days'], df['Cost'])
plt.title("Hospital Stay Duration vs Treatment Cost")
plt.xlabel("Days in Hospital")
plt.ylabel("Cost")
plt.show()
```

```python
plt.figure()
df['Insurance'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Insurance Coverage Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure()
plt.hist(df['Days'], bins=15)
plt.title("Distribution of Hospital Stay Duration")
plt.xlabel("Days")
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=pd.cut(df['Age'], bins=[0,20,40,60,80,100]), y=df['Cost'])
plt.title("Age Group vs Treatment Cost")
plt.xlabel("Age Group")
plt.ylabel("Cost")
plt.show()
```

```python
plt.figure()
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Stress_Level', hue='Disease', data=df)
plt.title("Stress Level vs Disease Type")
plt.xlabel("Stress Level")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.countplot(x='Smoke', hue='Disease', data=df[df['Disease']=='Heart Disease'])
plt.title("Smoking Habit in Heart Disease Patients")
plt.xlabel("Smoking")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.hist(df['Sugar'], bins=20)
plt.title("Blood Sugar Level Distribution")
plt.xlabel("Sugar Level")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Readmitted', y='Visits', data=df)
plt.title("Hospital Visits vs Readmission")
plt.xlabel("Readmitted")
plt.ylabel("Number of Visits")
plt.show()
```

```python
plt.figure()
df['Outcome'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Patient Outcome Distribution")
plt.ylabel("")
plt.show()
```
